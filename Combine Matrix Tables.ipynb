```python
import hail as hl
hl.init()
```

```python
data = hl.read_matrix_table("gs://pgr-wes-30k-mt/Vep_MT/Promis30_chr1_VEP.mt")
print("Chromosome 1", data.count())
```

```python
data.rows().show()
```

```python
for file in list(range(2,23))+['X','Y']:
    data1 = hl.read_matrix_table("gs://pgr-wes-30k-mt/Vep_MT/Promis30_chr" + str(file) + "_VEP.mt")
    print("Chromosome", file, data1.count())
    data = data.union_rows(data1)
```

```python
data.write("gs://pgr-wes-30k-mt/Vep_MT/Promis30_combined.mt")
```
